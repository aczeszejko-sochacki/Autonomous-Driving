```python
%matplotlib inline
from skimage.filters import roberts, sobel
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
```

```python
img = plt.imread('data/record_image.jpg')
```

```python
plt.imshow(img)
```

```python
# Need to convert image
img_gray = rgb2gray(img)
```

```python
plt.imshow(sobel(img_gray))
```

```python
plt.imshow(roberts(img_gray))
```
